```python
from collections import defaultdict

def can_finish_courses(N, prerequisites):
    graph = defaultdict(list)
    indegree = [0] * (N + 1)

    for prerequisite in prerequisites:
        graph[prerequisite[0]].append(prerequisite[1])
        indegree[prerequisite[1]] += 1

    courses_with_no_prerequisites = [course for course in range(1, N + 1) if indegree[course] == 0]
    order = []

    while courses_with_no_prerequisites:
        course = courses_with_no_prerequisites.pop(0)
        order.append(course)

        for neighbor in graph[course]:
            indegree[neighbor] -= 1
            if indegree[neighbor] == 0:
                courses_with_no_prerequisites.append(neighbor)

    if len(order) != N:
        return "IMPOSSIBLE"

    return " ".join(map(str, order))

with open('input1(b).txt', 'r') as file:
    lines = file.readlines()
    N, M = map(int, lines[0].split())
    prerequisites = []
    for line in lines[1:]:
        A, B = map(int, line.split())
        prerequisites.append((A, B))

result = can_finish_courses(N, prerequisites)

with open('output1(b).txt', 'w') as file:
    file.write(result + '\n')
```
