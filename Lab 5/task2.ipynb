```python
from collections import defaultdict, deque

def bfs_lexicographically_smallest(graph, in_degree, num_courses):
    queue = deque()
    order = []

    for course in range(1, num_courses + 1):
        if in_degree[course] == 0:
            queue.append(course)

    while queue:
        queue = deque(sorted(queue))
        current_course = queue.popleft()
        order.append(current_course)

        for next_course in graph[current_course]:
            in_degree[next_course] -= 1
            if in_degree[next_course] == 0:
                queue.append(next_course)

    return order if len(order) == num_courses else []

inpt2= open("input2.txt", "r")
outpt2= open("output2.txt", "w")

N, M = map(int, inpt2.readline().split())
graph = defaultdict(list)
in_degree = [0] * (N + 1)

for _ in range(M):
    A, B = map(int, inpt2.readline().split())
    graph[A].append(B)
    in_degree[B] += 1

course_order = bfs_lexicographically_smallest(graph, in_degree, N)

if not course_order:
    outpt2.write("IMPOSSIBLE")
else:
    for course in course_order:
        outpt2.write(str(course) + " ")
inpt2.close()
outpt2.close()
```
