```python
def DFS(graph, start):
    visited = [False] * (len(graph) + 1)
    in_degree = [0] * (len(graph) + 1)

    for node in graph:
        for neighbor in graph[node]["Adj"]:
            in_degree[neighbor] += 1

    stack = []
    for node in start:
        if in_degree[node] == 0:
            stack.append(node)
            visited[node] = True

    sorted_nodes = []
    while stack:
        node = stack.pop()
        sorted_nodes.append(node)

        for neighbor in graph[node]["Adj"]:
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0 and not visited[neighbor]:
                stack.append(neighbor)
                visited[neighbor] = True
    return sorted_nodes

infile = open("input1(a).txt")
outfile = open("output1(a).txt", "w")

a, b = map(int, infile.readline().strip().split())
graph = {}
node_list = []

for i in range(1, a + 1):
    node_list.append(i)
    graph[i] = {"Adj": [], "Indgree": 0}

for i in range(b):
    x, y = map(int, infile.readline().strip().split())
    graph[x]["Adj"].append(y)
    graph[y]["Indgree"] += 1

k = []
l = []
for i in range(len(node_list)):
    if graph[node_list[i]]["Indgree"] == 0:
        k.append(node_list[i])
    else:
        l.append(node_list[i])

node_list = k + l
new_l = DFS(graph, k)
if len(new_l) == len(node_list):
    s = " ".join(map(str, new_l))
    print(s, file=outfile)
else:
    print("IMPOSSIBLE", file=outfile)

infile.close()
outfile.close()
```
