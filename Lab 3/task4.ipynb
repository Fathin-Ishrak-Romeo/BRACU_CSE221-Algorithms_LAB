```python
file = open("input4.txt", "r")
out = open("output4.txt", "w")

size = int(file.readline())
lst = file.readline().split()

lst = [int(x) for x in lst if x]

def maxx(a, b):
    if a > b:
        return a
    else:
        return b

def divide(arr):
    if len(arr) <= 1:
        return arr[0] ** 2
    else:
        mid = len(arr) // 2
        a1 = divide(arr[:mid])
        a2 = divide(arr[mid:])
        return maxx(a1, a2)

result = None
large = None

for i in range(len(lst) - 1):
    ai = lst[i]
    aj2 = divide(lst[i + 1:])
    if large is None:
        large = ai
    elif ai > large:
        large = ai

    if result is None:
        result = ai + aj2
    else:
        if ai + aj2 > result:
            result = ai + aj2

out.write(f"{result}")
file.close()
out.close()
```
