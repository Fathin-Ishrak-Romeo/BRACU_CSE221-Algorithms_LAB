```python
def findMax(arr, left, right):
    if left == right:
        return arr[left]

    mid = (left + right) // 2
    max_left = findMax(arr, left, mid)
    max_right = findMax(arr, mid + 1, right)

    return max(max_left, max_right)

input_file_name = "input2.txt"
with open(input_file_name, "r") as input_file:
    n = int(input_file.readline().strip())
    numbers = list(map(int, input_file.readline().strip().split()))

max_value = findMax(numbers, 0, n - 1)

output_file_name = "output2.txt"
with open(output_file_name, "w") as output_file:
    output_file.write(str(max_value))


# The time complexity of this code is O(n) because it recursively divides the list into two halves
# and computes the maximum value for each half, resulting in a linear number of comparisons.
```
