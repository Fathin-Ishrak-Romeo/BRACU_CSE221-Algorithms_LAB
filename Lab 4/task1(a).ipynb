```python
def create_adjacency_matrix(N, M, edges):
    adjacency_matrix = [[0] * (N + 1) for _ in range(N + 1)]
    for edge in edges:
        u, v, w = edge
        adjacency_matrix[u][v] = w

    return adjacency_matrix

with open('input1(a).txt', 'r') as file:
    N, M = map(int, file.readline().split())
    edges = [list(map(int, file.readline().split())) for _ in range(M)]

adj_matrix = create_adjacency_matrix(N, M, edges)

with open('output1(a).txt', 'w') as file:
    for row in adj_matrix:
        file.write(' '.join(map(str, row)) + '\n')
```
