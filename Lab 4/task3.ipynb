```python
from collections import defaultdict

def dfs(graph, start, visited, traversal_order):
    visited.add(start)
    traversal_order.append(start)
    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited, traversal_order)

def main():
    with open('input3.txt', 'r') as file_in:
        N, M = map(int, file_in.readline().split())
        graph = defaultdict(list)
        for _ in range(M):
            u, v = map(int, file_in.readline().split())
            graph[u].append(v)
            graph[v].append(u)

    visited = set()
    traversal_order = []
    dfs(graph, 1, visited, traversal_order)

    with open('output3.txt', 'w') as file_out:
        file_out.write(' '.join(map(str, traversal_order)))

if __name__ == "__main__":
    main()
```
