```python
import numpy as np
input = open('/content/input6.txt', 'r')
output = open('/content/output6.txt', 'w')

a, b = list(map(int, input.readline().split()))
arr = np.zeros((a, b), dtype = int)
visited = np.zeros((a, b), dtype = bool)

for i in range(a):
  x = input.readline()
  for j in range(b):
    arr[i][j] = 0 if x[j] == '.' else 1 if x[j] == 'D' else -1
    visited[i][j] = False

def diamond(arr, a, b):
  count = 0
  def dfs(i, j):
    if not valid(i, j):
      return 0
    visited[i][j] = True
    total = 0
    if arr[i][j] == 1:
      total += 1
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    for u, v in directions:
      ip, jp = i + u, j + v
      total += dfs(ip, jp)
    return total

  def valid(i, j):
    return 0 <= i < a and 0 <= j < b and (arr[i][j] == 0 or arr[i][j] == 1) and not visited[i][j]
  for i in range(a):
    for j in range(b):
      if arr[i][j] == 0 and not visited[i][j]:
        count = max(count, dfs(i, j))
  return count

ans = diamond(arr, a, b)
output.write(str(ans))
input.close()
output.close()
```
