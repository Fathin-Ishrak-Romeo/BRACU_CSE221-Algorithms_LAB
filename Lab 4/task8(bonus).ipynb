```python
from collections import defaultdict

with open('input8.txt', 'r') as file_in, open('output8.txt', 'w') as file_out:
    T = int(file_in.readline().strip())
    for case in range(1, T + 1):
        n = int(file_in.readline().strip())
        rivals = defaultdict(int)
        for _ in range(n):
            u, v = map(int, file_in.readline().split())
            rivals[u] += 1
            rivals[v] += 1
        max_members = max(rivals.values())
        file_out.write(f"Case {case}: {max_members}\n")
```
