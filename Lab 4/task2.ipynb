```python
from collections import defaultdict, deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])
    visited.add(start)
    traversal_order = []

    while queue:
        node = queue.popleft()
        traversal_order.append(node)

        for neighbor in graph[node]:
            if neighbor not in visited:
                queue.append(neighbor)
                visited.add(neighbor)

    return traversal_order

def main():
    with open('input2.txt', 'r') as file_in:
        N, M = map(int, file_in.readline().split())
        graph = defaultdict(list)

        for _ in range(M):
            u, v = map(int, file_in.readline().split())
            graph[u].append(v)
            graph[v].append(u)

    bfs_order = bfs(graph, 1)

    with open('output2.txt', 'w') as file_out:
        file_out.write(' '.join(map(str, bfs_order)))

if __name__ == "__main__":
    main()
```
