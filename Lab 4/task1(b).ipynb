```python
def create_adjacency_matrix(N, edges):
    adjacency_matrix = {i: [] for i in range(1, N + 1)}

    for edge in edges:
        u, v, w = edge
        adjacency_matrix[u].append((v, w))
    return adjacency_matrix

def print_adjacency_matrix(N, adj_matrix):
    with open('output1(b).txt', 'w') as file:
        file.write('0 :\n')
        for i in range(1, N + 1):
            file.write(f"{i} : ")
            if i in adj_matrix:
                for v, w in adj_matrix[i]:
                    file.write(f"({v},{w}) ")
            file.write('\n')

with open('input1(b).txt', 'r') as file:
    N, M = map(int, file.readline().split())
    edges = [list(map(int, file.readline().split())) for _ in range(M)]

adj_matrix = create_adjacency_matrix(N, edges)

print_adjacency_matrix(N, adj_matrix)
```
