```python
from collections import defaultdict, deque

def shortest_path(graph, start, end):
    queue = deque([(start, [start])])
    visited = set()
    while queue:
        node, path = queue.popleft()
        visited.add(node)
        if node == end:
            return len(path) - 1, path
        for neighbor in graph[node]:
            if neighbor not in visited:
                queue.append((neighbor, path + [neighbor]))

with open('input5.txt', 'r') as file_in:
    N, M, D = map(int, file_in.readline().split())
    graph = defaultdict(list)
    for _ in range(M):
        u, v = map(int, file_in.readline().split())
        graph[u].append(v)
        graph[v].append(u)
    min_time, shortest_path = shortest_path(graph, 1, D)

with open('output5.txt', 'w') as file_out:
    file_out.write(f"Time: {min_time}\n")
    file_out.write(f"Shortest Path: {' '.join(map(str, shortest_path))}\n")
```
