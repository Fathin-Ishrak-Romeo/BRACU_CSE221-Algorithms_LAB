```python
def dfs(graph, start, visited):
    visited[start] = True
    count = 1

    for neighbor in graph[start]:
        if not visited[neighbor]:
            count += dfs(graph, neighbor, visited)

    return count

def find_maximum_cities(graph):
    max_cities = 0
    city_a, city_b = None, None

    for city in graph:
        visited = {node: False for node in graph}
        count = dfs(graph, city, visited)

        if count > max_cities:
            max_cities = count
            city_a, city_b = city, None

            for node, visit in visited.items():
                if visit:
                    city_b = node

    return city_a, city_b

with open('input7.txt', 'r') as file_in:
    N = int(file_in.readline().strip())
    graph = {i + 1: [] for i in range(N)}

    for _ in range(N - 1):
        u, v = map(int, file_in.readline().split())
        graph[u].append(v)
        graph[v].append(u)

city_A, city_B = find_maximum_cities(graph)

if city_A > city_B:
    city_A, city_B = city_B, city_A

with open('output7.txt', 'w') as file_out:
    file_out.write(f"{city_A} {city_B}\n")
```
