```python
from collections import defaultdict

def has_cycle(graph):
    visited = {node: False for node in graph}
    rec_stack = {node: False for node in graph}
    result = "NO"

    def has_cycle_util(node):
        nonlocal result
        visited[node] = True
        rec_stack[node] = True
        for neighbor in graph.get(node, []):
            if not visited.get(neighbor, False):
                if has_cycle_util(neighbor):
                    result = "YES"
                    return True
            elif rec_stack.get(neighbor, False):
                result = "YES"
                return True
        rec_stack[node] = False
        return False

    for node in graph:
        if result == "YES":
            break
        if not visited[node]:
            if has_cycle_util(node):
                break
    return result

with open('input4.txt', 'r') as file_in:
    N, M = map(int, file_in.readline().split())
    graph = defaultdict(list)

    for _ in range(M):
        u, v = map(int, file_in.readline().split())
        graph[u].append(v)
result = has_cycle(graph)

with open('output4.txt', 'w') as file_out:
    file_out.write(result)
```
