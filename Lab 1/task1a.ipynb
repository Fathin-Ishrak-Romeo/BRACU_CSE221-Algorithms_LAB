```python
def is_odd_or_even(n):
    if n % 2 == 0:
        return f"{n} is an Even number."
    else:
        return f"{n} is an Odd number."

with open("input1a.txt", "r") as input_file, open("output1a.txt", "w") as output_file:
    lines = input_file.readlines()

    t = int(lines[0].strip())
    for i in range(1, t + 1):
        n = int(lines[i].strip())
        result = is_odd_or_even(n)
        output_file.write(result + "\n")
```
