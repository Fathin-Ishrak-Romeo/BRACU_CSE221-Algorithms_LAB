```python
def calculate(expression):
    parts = expression.split()
    operand1 = float(parts[1])
    operator = parts[2]
    operand2 = float(parts[3])

    if operator == '+':
        result = operand1 + operand2
    elif operator == '-':
        result = operand1 - operand2
    elif operator == '*':
        result = operand1 * operand2
    elif operator == '/':
        if operand2 != 0:
            result = operand1 / operand2
        else:
            return "Undefined"
    return result

with open("input1b.txt", "r") as input_file:
    t = int(input_file.readline().strip())
    expressions = [input_file.readline().strip() for _ in range(t)]

with open("output1b.txt", "w") as output_file:
    for expression in expressions:
        result = calculate(expression)
        output_file.write(f"The result of {expression} is {result}\n")
```
