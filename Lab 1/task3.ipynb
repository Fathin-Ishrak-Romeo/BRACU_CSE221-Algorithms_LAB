```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i][1] > right_half[j][1] or (left_half[i][1] == right_half[j][1] and left_half[i][0] < right_half[j][0]):
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

with open("input3.txt", "r") as input_file:
    n = int(input_file.readline().strip())
    student_ids = list(map(int, input_file.readline().strip().split()))
    student_marks = list(map(int, input_file.readline().strip().split()))

students = [(student_ids[i], student_marks[i]) for i in range(n)]

merge_sort(students)

with open("output3.txt", "w") as output_file:
    for student in students:
        output_file.write(f'ID: {student[0]} Mark: {student[1]}\n')
```
