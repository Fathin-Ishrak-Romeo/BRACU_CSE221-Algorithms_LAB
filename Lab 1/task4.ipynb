```python
input_file_path = "input4.txt"
with open(input_file_path, "r") as input_file:
    n = int(input_file.readline())
    trains = [input_file.readline().strip() for _ in range(n)]

train_info = []
for train in trains:
    parts = train.split(" will departure for ")
    name = parts[0]
    time = parts[1].split(" at ")[1]
    train_info.append((name, time, train))

sorted_trains = sorted(train_info, key=lambda x: (x[0], x[1], trains.index(x[2])))
output_file_path = "output4.txt"

with open(output_file_path, "w") as output_file:
    for train in sorted_trains:
        output_file.write(train[2] + "\n")
```
