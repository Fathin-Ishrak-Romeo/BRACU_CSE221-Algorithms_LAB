```python
def find_pair_with_sum(N, S, numbers):
    for i in range(N):
        for j in range(i + 1, N):
            if numbers[i] + numbers[j] == S:
                return i + 1, j + 1
    return "IMPOSSIBLE"

with open("input1(1).txt", "r") as input_file:
    N, S = map(int, input_file.readline().split())
    numbers = list(map(int, input_file.readline().split()))

result = find_pair_with_sum(N, S, numbers)

with open("output1(1).txt", "w") as output_file:
    if result == "IMPOSSIBLE":
        output_file.write("IMPOSSIBLE\n")
    else:
        output_file.write(f"{result[0]} {result[1]}\n")
```
