```python
def max_tasks(N, tasks):
    tasks.sort(key=lambda task: task[1])
    completed_tasks = []
    end_time = -1

    for task in tasks:
        start, end = task
        if start >= end_time:
            completed_tasks.append(task)
            end_time = end

    return len(completed_tasks), completed_tasks

with open("input3.txt", "r") as input_file:
    N = int(input_file.readline())
    tasks = [list(map(int, input_file.readline().split())) for _ in range(N)]

max_tasks_count, completed_tasks = max_tasks(N, tasks)

with open("output3.txt", "w") as output_file:
    output_file.write(str(max_tasks_count) + "\n")
    for task in completed_tasks:
        output_file.write(" ".join(map(str, task)) + "\n")
```
