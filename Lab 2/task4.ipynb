```python
def max_activities(N, M, activities):
    # Sort activities by end time
    activities.sort(key=lambda x: x[1])

    assigned_tasks = 0
    end_times = [0] * M

    for activity in activities:
        start, end = activity
        assigned = False

        for i in range(M):
            if start >= end_times[i]:
                assigned_tasks += 1
                end_times[i] = end
                assigned = True
                break

        if assigned:
            continue

    return assigned_tasks

# Read input from "input4.txt" file
with open("input4.txt", "r") as input_file:
    N, M = map(int, input_file.readline().split())
    activities = [list(map(int, input_file.readline().split())) for _ in range(N)]

# Calculate the maximum number of activities that can be completed
result = max_activities(N, M, activities)

# Write the output to "output4.txt" file
with open("output4.txt", "w") as output_file:
    output_file.write(str(result) + "\n")
```
