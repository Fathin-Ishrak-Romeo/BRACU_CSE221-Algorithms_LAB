```python
f = open('input4.txt',"r")
f1 = open('output4.txt',"w")

n, e = map(int, f.readline().split())

t = 0
l = [[] for i in range (n + 1)]

r = []
for i in f:
  r+=[list(map(int, i.split()))]

for g, h, i in r :
      l[g] += [(h,i)]
      l[h] += [(g,i)]

s = [1]
while n > len(s):
      mct = None
      mm_cst = float('inf')
      for g in s:
          for h, i in l[g]:
              if i < mm_cst and h not in s:
                  mct = h
                  mm_cst = i
      t += mm_cst
      s += [mct]

f1.write(str(t))

f.close()
f1.close()
```
